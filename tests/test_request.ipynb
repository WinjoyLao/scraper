```python
from requests import Session
from bs4 import BeautifulSoup


TOPIC = 'https://www.zcool.com.cn/work/ZMjk3OTMxOTI=.html'
session = Session()
```

```python
WORK = 'https://www.zcool.com.cn/work/content/show?p={page}&objectId={objid}'
```

```python
resp = session.get(TOPIC)
```

```python
resp.text
```

```python
soup = BeautifulSoup(resp.text)
```

```python
element = soup.find('input', id='dataInput')
```

```python
objid = element.attrs.get('data-objid')
```

```python
WORK_URL = WORK.format(objid=objid, page=1)
work_resp = session.get(WORK_URL)
work_resp
```

```python
work_resp.json()
```

```python
from collections import namedtuple

Scrapy = namedtuple('Scrapy', 'type author title id index url')

data = work_resp.json().get('data', {})
author = data.get('product', {}).get('creatorObj', {}).get('username')
title = data.get('product', {}).get('title')
id = data.get('product', {}).get('id')
images = [Scrapy('page', author, title, id, obj.get('orderNo'), obj.get('url')) for  obj in data.get('allImageList')]
images
```
